```python
def test(pos, d, i, j, g):
    if (i, j) in d:
        return pos - d[(i, j)]
    elif i < 0 or j < 0 or i >= rows or j >= cols:
        return 0
    elif g[i][j] == '-':
        return 0
    elif g[i][j] == '>':
        g[i][j] = '-'
        d[(i, j)] = pos
        return test(pos+1, d, i, j+1, g)
    elif g[i][j] == '<':
        g[i][j] = '-'
        d[(i, j)] = pos
        return test(pos+1, d, i, j-1, g)
    elif g[i][j] == '^':
        g[i][j] = '-'
        d[(i, j)] = pos
        return test(pos+1, d, i-1, j, g)
    elif g[i][j] == 'V':
        g[i][j] = '-'
        d[(i, j)] = pos
        return test(pos+1, d, i+1, j, g)

answer = 0;
rows, cols = [int(x) for x in input().strip().split()]
grid = [[x for x in input().strip().split()] for _ in range(rows)]

for i in range(rows):
    for j in range(cols):
        if grid[i][j] != '-':
            answer = max(answer, test(0, dict(), i, j, grid))

print(answer)
```

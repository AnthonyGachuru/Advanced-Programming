```python
def plot_importances(clf,columns_used):
    """Plots the important variables used in the RandomForestClassifier to help you assess
    whether or not a variable should be kept in your data
    
    Parameters:
    clf (RandomForestClassifier) - A RandomForestClassifier that you have already fit to your data
    columns_used (list) - A list of columns that you used for input data
    
    Returns:
    None"""
    plt.figure(figsize=(10,12))
    ind = np.argsort(clf.feature_importances_)
    plt.barh(range(len(clf.feature_importances_)),clf.feature_importances_[ind])
    plt.title("Variable Importance")
    plt.yticks(range(len(clf.feature_importances_)),df.drop(['class'], axis=1).columns[ind])
    plt.show()
```

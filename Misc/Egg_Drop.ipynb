```python
import numpy as np

def find_floor(n,fail_floor):
    #Number of tests
    num_tests = 0
    #Initialize step and floors
    step_size = np.floor(np.sqrt(n))
    floor = step_size
    prev_floor = 1
    #Check whether we start too high
    if floor >= fail_floor:
        while prev_floor <= fail_floor:
            num_tests += 1
            prev_floor += 1
        return num_tests
    #If first floor succeeds
    else:
        #Test until first egg breaks
        while floor < fail_floor:
            prev_floor = floor
            floor += step_size
            num_tests += 1
        #Test until second egg breaks
        while prev_floor <= fail_floor:
            num_tests += 1
            prev_floor += 1
        return num_tests
    
floors = []
for i in xrange(10000):
    fail = np.random.randint(1,1000)
    floors.append(find_floor(1000,fail))
print np.mean(floors)
```
